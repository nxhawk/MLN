# Working with large amounts of data

```python
for index, df in enumerate(pd.read_csv("modified.csv", chunksize=5)):
    print(f"Chunk {index}:")
    df
    if index >= 3: break
```

```python

for index, dfs in enumerate(pd.read_csv("modified.csv", chunksize=5)):
    print(f"Chunk {index}:")
    dfs
    if index >= 3: break
```

### Read append dataframe with chunk

```python
df = pd.DataFrame()
for index, dfs in enumerate(pd.read_csv("modified.csv", chunksize=5)):
    print(f"Chunk {index}:")
    df = pd.concat([df, dfs])
    if index >= 3: break
df
```
